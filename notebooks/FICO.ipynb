# Data

## Data dictionary

| **Variable name**                    | **Description**                                                                                                                        | **Role**  |
|--------------------------------------|----------------------------------------------------------------------------------------------------------------------------------------|-----------|
| `RiskPerformance`                    | Paid as negotiated flag (12-36 Months). String of Good and Bad                                                                         | target    |
| `ExternalRiskEstimate`               | Consolidated version of risk markers                                                                                                   | predictor |
| `MSinceOldestTradeOpen`              | Months Since Oldest Trade Open                                                                                                         | predictor |
| `MSinceMostRecentTradeOpen`          | Months Since Most Recent Trade Open                                                                                                    | predictor |
| `AverageMInFile`                     | Average Months in File                                                                                                                 | predictor |
| `NumSatisfactoryTrades`              | Number Satisfactory Trades                                                                                                             | predictor |
| `NumTrades60Ever2DerogPubRec`        | Number Trades 60+ Ever                                                                                                                 | predictor |
| `NumTrades90Ever2DerogPubRec`        | Number Trades 90+ Ever                                                                                                                 | predictor |
| `PercentTradesNeverDelq`             | Percent Trades Never Delinquent                                                                                                        | predictor |
| `MSinceMostRecentDelq`               | Months Since Most Recent Delinquency                                                                                                   | predictor |
| `MaxDelq2PublicRecLast12M`           | Max Delq/Public Records Last 12 Months. See tab "MaxDelq" for each category                                                            | predictor |
| `MaxDelqEver`                        | Max Delinquency Ever. See tab "MaxDelq" for each category                                                                              | predictor |
| `NumTotalTrades`                     | Number of Total Trades (total number of credit accounts)                                                                               | predictor |
| `NumTradesOpeninLast12M`             | Number of Trades Open in Last 12 Months                                                                                                | predictor |
| `PercentInstallTrades`               | Percent Installment Trades                                                                                                             | predictor |
| `MSinceMostRecentInqexcl7days`       | Months Since Most Recent Inq excl 7days                                                                                                | predictor |
| `NumInqLast6M`                       | Number of Inq Last 6 Months                                                                                                            | predictor |
| `NumInqLast6Mexcl7days`              | Number of Inq Last 6 Months excl 7days. Excluding the last 7 days removes inquiries that are likely due to price comparision shopping. | predictor |
| `NetFractionRevolvingBurden`         | Net Fraction Revolving Burden. This is revolving balance divided by credit limit                                                       | predictor |
| `NetFractionInstallBurden`           | Net Fraction Installment Burden. This is installment balance divided by original loan amount                                           | predictor |
| `NumRevolvingTradesWBalance`         | Number Revolving Trades with Balance                                                                                                   | predictor |
| `NumInstallTradesWBalance`           | Number Installment Trades with Balance                                                                                                 | predictor |
| `NumBank2NatlTradesWHighUtilization` | Number Bank/Natl Trades w high utilization ratio                                                                                       | predictor |
| `PercentTradesWBalance`              | Percent Trades with Balance                                                                                                            | predictor |
